# 统计函数

```python
# 求平均值mean()

import numpy as np

ar1=np.arange(20).reshape(4,5)
print(ar1)
print(ar1.mean())
print(ar1.mean(axis=0)) # axis=0表示按行求平均值
print(ar1.mean(axis=1)) # axis=1表示按列求平均值
```

```python
# 中位数median(), 是按顺序排列的一组数据中居于中间位置的数. 若数组元素的个数为奇数,则中间元素就是中位数; 若为偶数, 中位数则为中间两个元素的平均值

ar1=np.array([1,3,5,6,8])
print(np.median(ar1)) # 数组中间的元素是5
ar2=np.array([1,3,5,6,8,9])
print(np.median(ar2)) # 数组中间两个元素的平均数为5.5
```

```python
# 标准差std()

# 在概率统计中最常使用作为统计分布程度上的测量, 是反映一级数据离散程度最常用的一种量化形式
# 定义: 是总体各单位标准值与其平均数离差平方的算术平均数的平方根. 即标准差是一级数据平均值分散程度的一种度量.
# 一个较大的标准差, 代表大部分数值是和其平均值之间的差异较大; 一个较小的标准差, 代表这些数值较接近平均值.

a=np.array([95,85,75,65,55,45])
b=np.array([73,72,71,69,68,67])

print(np.std(a))
print(np.std(b)) # b数组的标准差更小, 代表b数组元素更接近它的平均值
```

```python
# 方差var()

# 方差用于衡量一组数据的分散程度. 它是表示数据点与其均值之间的平均平方差.
# 它反映了数据的波动程度, 方差越大, 说明数据的分散程度越大; 反之, 说明数据的分散程度越小.
# 方差计算过程: 1.计算所有元素的平均值; 2.计算每个元素与平均值的差; 3.将每个平均值的差平方; 4.计算所有平方差的平均值.

ar1=np.array([1,2,3,4,5])
overall_var=np.var(ar1) # 计算总体方差
print("总体方差: ",overall_var)
simple_var=np.var(ar1,ddof=1) #计算样本方差
print("样本方差: ",simple_var)
```

```python
# 最大值max() / 最小值min() / 求和sum()

print(ar1)
print("最大值: ",ar1.max())
print("axis=0, 从上往下查找: ",ar1.max(axis=0))
print("axis=1, 从左往右查找: ",ar1.max(axis=1))
print("\n")
print("最小值: ",ar1.min())
print("axis=0, 从上往下查找: ",ar1.min(axis=0))
print("axis=1, 从左往右查找: ",ar1.min(axis=1))
print("\n")
print("总和: ",ar1.sum())
print("axis=0, 从上往下求和: ",ar1.sum(axis=0))
print("axis=1, 从左往右求和: ",ar1.sum(axis=1))
```

```python
# 加权平均值average(), 即将各数值乘以相应的权数, 得到总和, 再除以元素个数

ar1=[30,20,50]
print(np.average(ar1)) # 当没有权重时, 和np.mean()结果相同
print(np.mean(ar1))
```

  # 当必须设置skiprows参数时, 即使txt或csv文件前有注释行或空行, 也要将这些注释行或空行计算在内.

```python
# 读取普通文本文件

data=np.loadtxt('data/data1.txt', dtype=np.int8) # 不指定类型默认为float
print(data, data.shape)
```

```python
# 读取csv文件, csv默认分隔符是逗号, loadtxt中delimiter参数默认分隔符为空格或Tab

data=np.loadtxt('data/2020年1月份营收.csv', dtype=np.int32, delimiter=',', skiprows=20, usecols=range(1,4))
print(data,"\n",data.shape,"\n")
data=np.genfromtxt('data/2020年1月份营收.csv', dtype=None, delimiter=',', encoding='utf-8', skip_header=20, usecols=range(0,3))
print(data,"\n",data.shape,"\n")

# 用np.genfromtxt读取csv文档时, dtype=None可以让genfromtxt方法自动检测数据类型, 因此比loadtxt更灵活, 可以处理缺失值和混合数据类型
```

```python
dt=np.dtype([('姓名',np.str_,3),('年龄',np.int8),('性别',np.str_,1),('身高','i2')])
data=np.loadtxt('data/has_title.txt', dtype=dt, delimiter=' ', skiprows=1)
print(data,end="\n")
print("平均年龄: ", data['年龄'].mean(),end="\n")
print("身高中位数: ", np.median(data['身高']), end="\n")

# 计算所有女生的平均身高
isgirl=data['性别']=="女"
print(np.mean(data['身高'][isgirl]))
```

```python
def empty_to_zero(val):
    return 0 if val==b'' else int(val)

dt=np.dtype([('姓名',np.str_,3),('年龄',np.int8),('性别',np.str_,1),('身高','i2')])
data=np.loadtxt('data/has_title_empty.txt',
                dtype=dt,
                delimiter=' ',
                skiprows=1,
                converters={1:empty_to_zero, 3:empty_to_zero},
                usecols=(0,1,2,3))
print(data)
```
