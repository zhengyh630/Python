```python
import matplotlib as mpt
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-50,50)
y1=x**2
y2=x
plt.rcParams['font.sans-serif']=["SimHei"] # 用来设置字体样式，以正常显示中文标题
plt.rcParams['axes.unicode_minus']=False # 默认是使用Unicode负号，设置正常显示字符
plt.title('Y=X^2，X取值范围为：[-50,50)', fontsize=18)
plt.xlabel('X 轴', fontsize=12) # X轴名称
plt.ylabel('Y 轴', fontsize=12) # Y轴名称
plt.plot(x, y1, linewidth=5)
plt.plot(x, y2)
```

```python
x=np.arange(-50,50)
y1=x**2
y2=x
plt.rcParams['font.sans-serif']=["SimHei"] # 用来设置字体样式，以正常显示中文标题
plt.rcParams['axes.unicode_minus']=False # 默认是使用Unicode负号，设置正常显示字符
plt.title('Y=X^2，X取值范围为：[-50,50)', fontsize=18)

x_xticks=[-50,-40,-30,-20,-10,0,10,20,30,40,50]
x_labels=[f"{i}度" for i in x_xticks] # 将X轴的标签数字后面加上单位
plt.xticks(x_xticks, x_labels)
plt.plot(x, y1, linewidth=5)
plt.plot(x, y2)
```

```python
# 每个时间点的销量绘图

times = ['2015/06/26','2015/3/1','2015/9/6','2005/10/12','2015/11/7','2015/12/23','2016/1/28','2016/3/4','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']

sales=np.random.randint(500,2000,size=len(times))

plt.xticks(rotation=45)
plt.plot(times,sales) # times为X轴
```

```python
plt.xticks(range(1,len(times),2),labels=[1,2,3,4,5,6,7] ) # 通过labels改变X轴的标签。
plt.plot(times,sales) # times为X轴
```
