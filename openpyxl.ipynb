# 用openpyxl处理Excel文件

```python
from openpyxl import Workbook,load_workbook

wb=Workbook() #在内存中创建一个Excel
sheet=wb.active #读取活跃的sheet
sheet.title="第一个数据表"

# wb=load_workbook("excel_openpyxl.xlsx") #打开一个excel文件
```

写入数据

```python
#写入一个单元格中
sheet["A2"]="测试"
sheet["A4"]="第二条测试"

#从第一列最下方空白处,最左开始追加一行元素,一个元素占一个单元格
sheet.append(["zyh","男",50])

wb.save("excel_openpyxl.xlsx")
```

```python
#遍历整个sheet表

wb=load_workbook("巡查台账.xlsx")
sheet=wb["第一组"]
for row in sheet:
    for cell in row:
        if cell.value is None:
            print(" ")
        else:
            print(cell.value,end="\t")
    print("\n")
```

```python
# 遍历指定行或列，用sheet_iter指定最大和最小行或列
import datetime

wb=load_workbook("巡查台账.xlsx")
sheet=wb["第五组"]
for row in sheet.iter_rows(min_row=6): #按指定行遍历,若遍历所有行,则用sheet.rows
    for cell in row:
        if cell.value is None:
            print("",end="\t")
        elif isinstance(cell.value,datetime.datetime):
            print(cell.value.date(),end="\t")
        else:
            print(cell.value,end="\t")
    print("\n")
```

```python
wb=load_workbook("巡查台账.xlsx")
sheet=wb["第一组"]
for col in sheet.iter_cols(min_col=2,max_col=4): #按指定列遍历,若遍历所有列,则用sheet.columns
    for cell in col:
        if cell.value is None:
            print("",end="\t")
        elif isinstance(cell.value,datetime.datetime):
            print(cell.value.date(),end="\t")
        else:
            print(cell.value,end="\t")
    print("\n")
```

```python
from openpyxl import load_workbook
from openpyxl.styles import Font,Color,Alignment

wb=load_workbook("巡查台账.xlsx")
sheet=wb["第一组"]
font=Font(name="文鼎ＰＬ简中楷",size=20,bold=True,color="FF0000")
for col in sheet.iter_cols(min_col=3,max_col=3):
    for cell in col:
        cell.font=font
wb.save("巡查台账_test.xlsx")
```
