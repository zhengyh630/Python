```python
import matplotlib as mpt
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-50,51)
y=x**2

plt.plot(x,y) # 使用plt的plot()函数对x, y进行绘制线性图.
```

# matplotlib的基本方法

```python
# 图表名称plt.title()

# 设置中文字体  
font_path = '/usr/share/fonts/truetype/arphic-gkai00mp/gkai00mp.tff'  # 根据你的系统调整字体路径  
font_prop = fm.FontProperties(fname=font_path)
x=np.arange(-50,51)
y=x**2
plt.rcParams["font.sans-serif"]=["gkai00mp"]
plt.rcParams['axes.unicode_minus']=False
plt.title('y=x^2表', fontproperties=font_prop)
plt.plot(x,y)
```
